```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
try:
    data = pd.read_csv('https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto13/CasosNuevosCumulativo.csv')
except Exception as Error:
    print('Hubo un error al leer este dataset: ', Error)
```

```python
ultima_columna = data.iloc[: , -1:]
fecha_ultima_columna = ultima_columna.columns[0] # Obtener la ultima fecha
```

```python
# Regiones más importantes de país:
reg_Arica = data[fecha_ultima_columna][0]
reg_Valparaiso = data[fecha_ultima_columna][5]
reg_Metropolitana = data[fecha_ultima_columna][6]
reg_BioBio = data[fecha_ultima_columna][10]
total = data[fecha_ultima_columna][16]
```

```python
# Gráfica en plots

reg = ['Arica', 'Valparaiso', 'Metropolitana', 'BioBio', 'Casos Total']
var = [reg_Arica, reg_Valparaiso, reg_Metropolitana, reg_BioBio, total]

plt.bar(reg, var, color = 'orange')
plt.ylabel('Numero de casos')
plt.xlabel('Regiones')
plt.title('Casos nuevos por regiones')
```
